# IQR

```python
age_iqr = dt2.age.quantile(0.75) - dt2.age.quantile(0.25)
age_iqr
```

```python
edu_iqr = dt2['education.num'].quantile(0.75) - dt2['education.num'].quantile(0.25)
hrs_iqr = dt2['hours.per.week'].quantile(0.75) - dt2['hours.per.week'].quantile(0.25)
hrs_iqr
```

Wartość danych jest punktem oddalonym, jeśli:
a) jest położona przynajmniej o 1,5(IQR) poniżej Q1 lub
a) jest położona przynajmniej o 1,5(IQR) powyżej Q3

```python
# q1 - 1,5(iqr)
dt2.age.quantile(0.25) - 1.5 * age_iqr
```

```python
# q3 + 1,5(iqr)
dt2.age.quantile(0.75) + 1.5 * age_iqr
```
